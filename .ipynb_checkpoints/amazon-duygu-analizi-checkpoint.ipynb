```python
from warnings import filterwarnings
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL import Image
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, GridSearchCV, cross_validate
from sklearn.preprocessing import LabelEncoder
from textblob import Word, TextBlob
from wordcloud import WordCloud
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import nltk
```

```python
df = pd.read_excel("amazon.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Review'] = df['Review'].str.lower()
#Review sütunundaki tüm harfleri küçük harfe çevirdik.
```

```python
df['Review'] = df['Review'].str.replace('[^\w\s]', '', regex=True)
#Review sütunundaki noktalama işaretlerini yok ettik. 
```

```python
df['Review'] = df['Review'].str.replace('\d', '', regex=True)
#Review sütunundaki sayısal ifadeleri çıkardık.
```

```python
sw = stopwords.words('english')
```

```python
df['Review'] = df['Review'].apply(lambda x: " ".join(x for x in str(x).split() if x not in sw))
#Review sütununu taradık sw listesinde olan kelimeleri çıkardık.
```

```python
df['Review']
```

```python
temp_df = pd.Series(' '.join(df['Review']).split()).value_counts()
temp_df
#Nadir Kelimeler (RareWords) Listesi
```

```python
drops = temp_df[temp_df <= 1]
drops 
#Metinde 1 veya daha az geçen kelimeleri gruplandırdık ve adlandırdık. 'drops'
```

```python
df['Review'] = df['Review'].apply(lambda x: " ".join(x for x in x.split() if x not in drops))
#Review sütununda geçen nadir kelimeleri çıkarttık
```

```python
df["Review"].apply(lambda x: TextBlob(x).words).head() 
#Her bir kelimeyi tek tek seçti
```

```python
df['Review'] = df['Review'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
#Lemmatizasyon işlemi. Kelimelerin bulduğumuz köklerini aldı ve en yalın halleriyle değiştirdi.
```

```python
df['Review']
```

# Duygu Analizi

```python
sia = SentimentIntensityAnalyzer()
#Bu kodda bir sentimentAnalyzer nesnesi oluşturduk.
```

```python
df["Review"][0:10].apply(lambda x: sia.polarity_scores(x))
#Her satırdaki Review değişkenleri için duygu skorlarını hesaplar.
```

```python
df["Review"][0:10].apply(lambda x: sia.polarity_scores(x)["compound"])
#Duygu skorlarının compound değerlerini alır. 
```

```python
df["Review"].apply(lambda x: "pos" if sia.polarity_scores(x)["compound"] > 0 else "neg")
#eğer Compound değeri 0'dan büyükse pozitif değilse negatif olarak adlandırılır.
df["sentiment_label"] = df["Review"].apply(lambda x: "pos" if sia.polarity_scores(x)["compound"] > 0 else "neg")
#sentiment_label adında bir sütun eklenip pozitif veya negatif oluşu df'e eklenir.
```

```python
df.groupby("sentiment_label")["Star"].mean()
#Sentiment_label sütununu gruplayarak Star sütunun ortalamasını hesaplar.

#NOT:SentimentIntensityAnalyzer ile yorumları etiketleyerek, 
#yorum sınıflandırma makine öğrenmesi modeli için bağımlı değişken oluşturulmuş oldu.
```

# Makine Öğrenmesine Hazırlık


```python
train_x, test_x, train_y, test_y = train_test_split(df["Review"],
                                                    df["sentiment_label"],
                                                    random_state=42)
# Datayı train ve test data olarak ayırdık
```

```python
tfidfvect = TfidfVectorizer().fit(train_x)
#TfidfVectorizer kullanarak bir nesne oluşturduk ve train datamızı fit ettik
```

```python
x_train_tf_idf_word = tfidfvect.transform(train_x)
x_test_tf_idf_word = tfidfvect.transform(test_x)

#Oluşturduğumuz vektörleri train ve test datalarına transform ettik. 
```

# Modelleme (Lojistik Regresyon)

```python
log_model = LogisticRegression().fit(x_train_tf_idf_word, train_y)
#Lojistik regresyon modelini kurarak train dataları ile fit ettik
```

```python
y_pred=log_model.predict(x_test_tf_idf_word)
# Predict fonksiyonu ile test datasını tahmin ettik
```

```python
print(classification_report(test_y, y_pred))
# Modelin Performans tahmin sonuçları:
```

```python
cross_val_score(log_model, x_test_tf_idf_word, test_y, cv=5).mean()
# cross validation kullanarak ortalama accuracy (doğruluk) değerini hesapladık
```

```python
orneklem=pd.Series(df['Review'].sample(1).values)
# Sample fonksiyonunu kullanarak Review içinden bir örneklem seçtik
```

```python
vectorizer = CountVectorizer()
yeni_yorum = CountVectorizer().fit(train_x).transform(orneklem)
# Vektörleştirdiğim örneklemi fit ve transform ettim. Yeni gelen yorumu modele öğretip sayısal değere dönüştürdüm.
pred = log_model.predict(yeni_yorum)
#Log_model benim modelim. Burada da modele yeni gelen yorumu verdim. Tahmin etti. 
print(f'Review:  {orneklem[0]} \n Prediction: {pred}')
```

# Modelleme

```python
rf_model = RandomForestClassifier().fit(x_train_tf_idf_word, train_y)
#Random forest classifier kullanarak tahmin sonuçlarını hesapladık burada da 
```

```python
cross_val_score(rf_model, x_test_tf_idf_word, test_y, cv=5, n_jobs=-1).mean()
#Test verisinin sonuçları
```
